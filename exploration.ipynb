```python
df2 = pd.read_csv("archive/mimic-iv-clinical-database-demo-2.2/hosp/d_labitems.csv")
```

```python

# Merge on itemid
labs = df.merge(
    df2,
    on="itemid",
    how="left"
)

# Sort by patient + time
labs = labs.sort_values(["subject_id", "charttime"])


```

```python

# Inspect
labs_filtered = labs.dropna(subset=["value"])
labs_filtered.head(100)
```
